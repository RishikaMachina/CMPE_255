## Starting with the Model Application. Here we will be looking at 3 different Machine learning algorithms



### Naive Bayes on Reviews 

```python
#Train Naive Bayes - On CSR matrix
# Step 1 - Importing libraries
from sklearn.naive_bayes import MultinomialNB
from sklearn import metrics
from sklearn.preprocessing import LabelEncoder
```

```python
# Step 2 - Training and application of Model
nbModel = MultinomialNB().fit(X_train_idf, yTrain)
df = pickle.dump(nbModel, open("nbModel.pkl", 'wb'))

# Step 3 - Making Predictions
predicted = nbModel.predict(X_test_idf)
print(metrics.classification_report(yTest, predicted))
```

### XGBoost Algorithm on Reviews

As the Dataset is highly imbalanced so next plausible choice is XGBoost, which works well with imbalanced DataSet 

```python
#Train using XGBoost on csr matrix
# Step 1 - Importing libraries
from xgboost import XGBClassifier
```

```python
# Step 2 - Train and fit model ontraining data
modelXGB = XGBClassifier(verbosity=3, random_state=42)
modelXGB.fit(X_train_idf, yTrain)

# Step 3 - Make predictions for test data
y_pred = modelXGB.predict(X_test_idf)
print(metrics.classification_report(yTest, y_pred))
```
